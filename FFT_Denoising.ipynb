## **Import**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_pickle('jizoh.strain.2020-09-18.pkl')
df.head(10)
```

## **Function**

```python
channel = ["IN_S1_C", "IN_S2_C", "W_G2_C", "W_G3_C", "OUT_S1_A1", "OUT_S2_A1", "OUT_S3_A1", "R_G1_A1", "R_G2_A1", "R_G3_A1", "R_G4_A1"]
time_step = 1/200
sample_rate = 200   # number of samples taken per second
```

```python
def time_domain(channel, start_time, end_time):
    t = np.arange(start_time, end_time, time_step)  
    signal = df[channel][(start_time*sample_rate):(end_time*sample_rate)]

    fig = plt.figure(figsize=(9, 3), dpi=150) 
    plt.title(channel)
    plt.plot(t, signal, color = 'tab:red', LineWidth=1.5, label='Noisy Signal')
    plt.xlabel('Time(s)')
    plt.ylabel('Amplitude')
    plt.xlim(start_time, end_time)
    plt.legend()
    plt.show()
```

```python
def frequency_domain(channel, start_time, end_time):
    t = np.arange(start_time, end_time, time_step)  
    signal = df[channel][(start_time*sample_rate):(end_time*sample_rate)]

    n = len(t)
    fhat = np.fft.fft(signal,n)
    PSD = fhat * np.conj(fhat) / n
    freq = 1/(time_step*n) * np.arange(n)
    L = np.arange(1, np.floor(n/2), dtype='int')

    fig = plt.figure(figsize=(9, 3), dpi=150) 
    plt.title(channel)
    plt.plot(freq[L], PSD[L], color='tab:red', LineWidth=1.5, label='Noisy Signal')
    plt.xlabel('Frequency(Hz)')
    plt.ylabel('Power')
    plt.xlim(freq[L[0]], freq[L[-1]])
    plt.legend()
    plt.show()
```

```python
def fft_denoising(channel, start_time, end_time, thres_power):
    t = np.arange(start_time, end_time, time_step)  
    signal = df[channel][(start_time*sample_rate):(end_time*sample_rate)]

    n = len(t)
    fhat = np.fft.fft(signal, n)
    PSD = fhat * np.conj(fhat) / n
    freq = 1/(time_step*n) * np.arange(n)
    L = np.arange(1, np.floor(n/2), dtype='int')

    indices = PSD >= thres_power
    PSDclean = PSD * indices  
    fhat = indices * fhat
    ffilt = np.fft.ifft(fhat)

    fig = plt.figure(figsize=(9, 3), dpi=150) 
    plt.title(channel)
    plt.plot(t, signal, color='tab:red', LineWidth=1.5, label='Noisy Signal')
    plt.plot(t, ffilt, color='tab:cyan', LineWidth=1.5, label='Filtered at thres_power = %s'%str(thres_power))
    plt.xlabel('Time(s)')
    plt.ylabel('Amplitude')
    plt.xlim(start_time, end_time)
    plt.legend()
    plt.show()

    fig = plt.figure(figsize=(9, 3), dpi=150) 
    plt.title(channel)
    plt.plot(freq[L], PSD[L], color='tab:red', LineWidth=1.5, label='Noisy Signal')
    plt.plot(freq[L], PSDclean[L], color='tab:cyan', LineWidth=1.5, label='Filtered at thres_power = %s'%str(thres_power))
    plt.xlabel('Frequency(Hz)')
    plt.ylabel('Power')
    plt.xlim(freq[L[0]], freq[L[-1]])
    plt.legend()
    plt.show()
```

## **Main**

```python
with plt.style.context('dark_background'): 
    time_domain(channel[0],0,10)
    frequency_domain(channel[0],0,10)
```

```python
with plt.style.context('dark_background'): fft_denoising(channel[0],0,10,100)
```
